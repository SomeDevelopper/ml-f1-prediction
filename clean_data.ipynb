```python
import pandas as pd
from category_encoders import OrdinalEncoder 

csv_name_season = 'season_data.csv'
#csv_name_nor = 'nor-extended-data.csv'

df = pd.read_csv(csv_name_season)

df = df.sort_values(by=["Year", "Driver", "LapNumber"])
```

```python
df
```

```python
map_enc = [{
    "col": "Compound",
    "mapping": {
        'SOFT': 1,
        'MEDIUM': 2,
        'HARD': 3,
        'INTERMEDIATE': 4,
        'WET': 5
    }}]

encoder = OrdinalEncoder(mapping=map_enc)
df['Compound'] = encoder.fit_transform(df[['Compound']])
df['PitStopNextLap'] = 0

```

```python
# map_enc_driver = [{
#     "col": "Driver",
#     "mapping": {
#         'ALB': 1,
#         'BOT': 2,
#         'ALO': 3,
#         'GAS': 4,
#         'HAM': 5,
#         'HUL': 6,
#         'LAW': 7,
#         'LEC': 8,
#         'MAG': 9,
#         'NOR': 10,
#         'OCO': 11,
#         'PER': 12,
#         'PIA': 13,
#         'RUS': 14,
#         'SAI': 15,
#         'SAR': 16,
#         'STR': 17,
#         'TSU': 18,
#         'VER': 19,
#         'ZHO': 20
#     }}]

# encoder = OrdinalEncoder(mapping=map_enc_driver)
# df['Driver'] = encoder.fit_transform(df[['Driver']])

df = df.sort_values(by=["Year", "Driver", "LapNumber"]).reset_index(drop=True)
```

```python
for i in range(1, len(df)):
    same_driver = df.loc[i, "Driver"] == df.loc[i-1, "Driver"]

    compound_changer = df.loc[i, "Compound"] != df.loc[i-1, "Compound"]
    stint_increased = df.loc[i, "Stint"] > df.loc[i-1, "Stint"]

    if same_driver and (compound_changer or stint_increased):
        lap_pitsop = df.loc[i, "LapNumber"]

        df.loc[(df['Driver'] == df.loc[i, "Driver"]) & 
        (df['LapNumber'].between(lap_pitsop - 3, lap_pitsop - 1)), 'PitStopNextLap'
        ] = 1

```

```python
df['TotalLaps'] = df.groupby(['Driver'])['LapNumber'].transform('max')

df['LapsRemaining'] = df['TotalLaps'] - df['LapNumber']

df.drop(columns=['TotalLaps'], inplace=True)
```

```python
time_columns = ["LapTime", "Sector1Time", "Sector2Time", "Sector3Time"]

def convert_time_columns(time_str):
    try:
        if pd.isna(time_str) or isinstance(time_str, float):
            return time_str
        h, m, s = map(float, time_str.replace(" days ", ":").split(":")[-3:])
        return h * 3600 + m * 60 + s
    except:
        return None

for col in time_columns:
    df[col] = df[col].apply(convert_time_columns)

df
```

```python
# Replace True and False with 1 and 0
if df['FreshTyre'].dtype == bool:
    df['FreshTyre'] = df['FreshTyre'].astype(int)
else:
    df['FreshTyre'] = df['FreshTyre'].map({'True': 1, 'False': 0})
df
```

```python
# Drop rows with missing target

df = df.dropna(subset=['Position'])
df = df.apply(lambda x: x.fillna(x.median()) if x.dtype in ['float64', 'int64'] else x)
```

```python
# strandardisation des données

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

cols_to_scale = ["LapTime", "Sector1Time", "Sector2Time", "Sector3Time"]

df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])

```

```python

df.to_csv("season_data_preprocessed.csv", index=False)
#df.to_csv("nor-extended-data-preprocessed.csv", index=False)

print("✅ Normalisation terminée et dataset sauvegardé.")
```
